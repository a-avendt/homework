```python
##Q2
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5, 6, 7, 8, 9, 10])

plt.plot(x, y)
plt.show()
```

```python
##Q5
marks = {"Andy":88, "Amy":66, "James":90, "Jules":55, "Arthur":77}
x = marks.keys()
y = marks.values()

plt.title("Student Grades")
plt.xlabel("Student Name")
plt.ylabel("Grade")

plt.bar(x, y)
```

```python
##pie chart
plt.pie(y, labels = x)
plt.title("Student Grades")
plt.legend(title = "Student Name", bbox_to_anchor=(1.05, 1.0), loc = 'upper left')
plt.show()
```

```python
##Q6
##multiple lines
x = np.array([1, 2, 3, 4])
y1 = np.array([1, 2, 4, 8])
y2 = np.array([1, 3, 9, 27])
y3 = np.array([1, 4, 16, 64])

plt.subplot(3, 2, 1)
plt.tight_layout()
plt.plot(x, y1, x, y2, x, y3)
plt.title("Multiple Lines Plot")

##bar chart
plt.subplot(3, 2, 2)
plt.tight_layout()
plt.bar(x, y1)
plt.title("Bar Chart")

##pie chart
pies = ("Pecan", "Apple", "Key Lime", "Pumpkin")
plt.subplot(3, 2, 3)
plt.tight_layout()
plt.pie(y1, labels = pies)
plt.title("Pie Chart")


##grid
plt.subplot(3, 2, 4)
plt.tight_layout()
plt.plot(x, y1, x, y2)
plt.grid()
plt.title("Grid Plot")

##scatter plot
plt.subplot(3, 2, 5)
plt.tight_layout()
plt.scatter(x, y1)
plt.title("Scatter Plot")

##histogram plot
plt.subplot(3, 2, 6)
plt.tight_layout()
x2 = np.random.normal(0, 10, 200)
plt.hist(x2)
plt.title("Histogram")
```
