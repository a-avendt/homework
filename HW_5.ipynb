```r
#Q2
my_data_frame = data.frame (
    ID = c(1, 2, 3, 4, 5),
    Name = c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
    Salary = c(623.30, 515.20, 611.00, 729.00, 843.25)
)

my_data_frame

#part b
add_column = cbind(my_data_frame, Department = c("IT", "Finance", "Computer Science", 
                                                 "Sales", "Customer Service"))

add_column

#part c
extract_df = add_column[-c(1, 3, 5), -c(2, 3)]

extract_df

#part d (bar chart)
library(ggplot2)
extract_df1 = add_column[c(1, 4, 5), c(2, 3)]


require(ggplot2)
barx = ggplot(data = extract_df1, aes(extract_df1$Name, extract_df1$Salary)) + 
    geom_bar(stat = "identity") + xlab("Name") + ylab("Salary") +ggtitle("Salaries")

#rotating the axis
barx + theme(axis.text.x = element_text(ang = 90))



#part e (pie chart)
max_sal = max(add_column$Salary)
min_sal = min(add_column$Salary)
mean_sal = mean(add_column$Salary)
x = c(max_sal, min_sal, mean_sal)
mylabel = c("Highest Salary", "Lowest Salary", "Average Salary")
pie(x, label = mylabel, main = "Salaries")






                   
```

```r
#Q3
ff = read.csv('amazon-orders.csv')
head(ff) #this gets the first 5 rows, and just lets me check to make sure I've done things right


#cleaning the data--replacing NaN with 0
ff[is.na(ff)] = 0

#checking
head(ff)

#summary of total charged and tax charged
ff$Total.Charged = as.numeric(gsub("\\$", "", ff$Total.Charged))
head(ff)

summary(ff$Total.Charged)

ff$Tax.Charged = as.numeric(gsub("\\$", "", ff$Tax.Charged))

summary(ff$Tax.Charged)

#ff$Order.Date = as.Date(ff$Order.Date, '%m/%d/%Y')
head(ff)
```

```r
library(ggplot2)
```

```r
#bar chart of order total by date
require(ggplot2)
bar = ggplot( data = ff, aes(Order.Date, Total.Charged)) + geom_bar(stat = "identity")

#rotating the axis
bar + theme(axis.text.x = element_text(ang = 90))
```

```r
library(dplyr)
```

```r
#grouping by month
print("Monthly Totals:")
library(lubridate)
newOrder.Date = as.Date(ff$Order.Date, '%m/%d/%Y')

monthly_total = ff %>% group_by(month = floor_date(newOrder.Date, "month")) %>%
        summarize(Total.Charged = sum(Total.Charged))
monthly_total

print("Minimum Purchase Each Month")
monthly_min = ff %>% group_by(month = floor_date(newOrder.Date, "month")) %>%
        summarize(Total.Charged = min(Total.Charged))
monthly_min

print("Maximum Purchase Each Month")
monthly_max = ff %>% group_by(month = floor_date(newOrder.Date, "month")) %>%
        summarize(Total.Charged = max(Total.Charged))
monthly_max

print("Average Purchase Each Month")
monthly_avg = ff %>% group_by(month = floor_date(newOrder.Date, "month")) %>%
        summarize(Total.Charged = mean(Total.Charged))
monthly_avg

print("Monthly Standard Deviation of Purchase Amount")
monthly_stdev = ff %>% group_by(month = floor_date(newOrder.Date, "month")) %>%
        summarize(Total.Charged = sd(Total.Charged))
monthly_stdev

print("Aggregate Statistics")
summary(ff$Total.Charged)
```

```r
#pie chart
lbl = c("April", "June", "July", "August")
pie(monthly_total$Total.Charged, labels = lbl, main = "Amazon Spending by Month")
```

```r
#spending over time


plot1 = ggplot(data = ff, aes(x=newOrder.Date, y=Total.Charged)) + geom_line() +geom_point()

#rotating the axis
plot1 + theme(axis.text.x = element_text(ang = 90)) +labs(title = "Daily Tax Amount") + xlab("Order Date")
```

```r
#multiple line charts

#daily purchase and tax amount
colors1 = c("Total" = "red", "Tax" = "blue")
ggplot(data = ff, aes(x = newOrder.Date)) +
    geom_line(aes(y = Total.Charged, color = "Total")) +
    geom_line(aes(y = Tax.Charged, color = "Tax")) + 
    labs(x = "Order Date", y = "Amount", color = "Legend") +
    scale_color_manual(values = colors1) + ggtitle("Daily Total and Tax Amounts")
```
